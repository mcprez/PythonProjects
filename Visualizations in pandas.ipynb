# Visualizations in pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Marc\Desktop\Data Analyst Apps & Filles & projects\Panda files\Ice Cream Ratings.csv")
df
```

```python
print(plt.style.available)
plt.style.use('dark_background')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores' )
```

```python
print(plt.style.available)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating', xlabel = 'Daily Rating', ylabel = 'Scores', subplots = True)
```

```python
df.plot(kind = 'bar', title = 'Ice cream Rating', ylabel = 'Scores'
       , stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', title = 'Ice cream Rating', ylabel = 'Scores'
       , stacked = True)
```

```python
df.plot.barh(title = 'Ice cream Rating', ylabel = 'Daily Rating', xlabel = 'Scores'
       , stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', title = 'Ice Cream Rating', s = 100, c = 'Red')
```

```python
df.plot.hist(bins = 20, title = 'Ice Cream Rating')
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,5), title = 'Ice Cream Rating')
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10,6), title = 'Ice Cream Rating')
```
